```python
import pandas as pd
import random
from tqdm import tqdm
import numpy as np

# 1. Define feature space
features = {
    "AgeGroup": ["18–29", "30–44", "45+"],
    "HairLoss": ["Yes", "No"],
    "Dandruff": ["Yes", "No"],
    "HairGreying": ["Yes", "No"],
    "HL_Duration": ["<3 months", "3–6 months", ">6 months"],
    "HL_Shedding": ["Rarely", "Occasionally", "Frequently"],
    "DF_Type": ["Powdery", "Flaky"],
    "DF_ItchyScalp": ["Rarely", "Frequently"],
    "GR_Duration": ["<3 months", "<1 year", "<1.5 years"],
    "GR_TreatmentType": ["Coloring/Keratin", "None"],
    "RegularUsage": ["Yes", "No"],
    "WashFrequency": ["Rarely", "2–3 times/week", "Frequently"],
    "TakesSupplements": ["Yes", "No"],
    "StressLevel": ["Rarely", "Occasionally", "Frequently"],
    "FamilyHistory": ["Yes", "No"]
}

# 2. Sample generation
def generate_user_response():
    response = {key: random.choice(val) for key, val in features.items()}
    return response

# 3. Product mapping based on decision logic
def map_products(user):
    # Primary HL
    if user["HairLoss"] == "Yes":
        if user["HL_Duration"] == "<3 months":
            primary_hl = "Redensyl + Biotin Combo Serum"
            support_hl = "Shampoo + Conditioner for Hairfall"
        elif user["HL_Duration"] == "3–6 months":
            primary_hl = "Redensyl + Biotin + Mild Minoxidil (Pilgrim, Be Bodywise)"
            support_hl = "DHT Blocker Shampoo + Conditioner"
        else:
            primary_hl = "Minoxidil 5% Foam + Redensyl + Microneedling"
            support_hl = "DHT Blocker + Supplements"
        lifestyle_hl = "Scalp Massage, Diet Improvement"

    else:
        primary_hl = support_hl = lifestyle_hl = "Not Applicable"

    # Primary DF
    if user["Dandruff"] == "Yes":
        if user["DF_Type"] == "Powdery" and user["DF_ItchyScalp"] == "Frequently":
            primary_df = "Ketoconazole + Salicylic Acid Shampoo"
            support_df = "Tea Tree Scalp Serum"
        else:
            primary_df = "ZPTO + pH Balancing Anti-Dandruff Shampoo"
            support_df = "Aloe Vera + Niacinamide Serum"
        lifestyle_df = "Avoid Hot Water, Gentle Shampoo"

    else:
        primary_df = support_df = lifestyle_df = "Not Applicable"

    # Primary GR
    if user["HairGreying"] == "Yes":
        if user["GR_TreatmentType"] == "Coloring/Keratin":
            primary_gr = "Hair Strengthening Oil + Amla Tablets"
        elif user["GR_Duration"] == "<3 months":
            primary_gr = "Vitamin B12 + Biotin Gummies"
        elif user["GR_Duration"] == "<1 year":
            primary_gr = "Melanin Boost Hair Oil"
        else:
            primary_gr = "Anti-Grey Hair Serum + Vitamin Supplements"
    else:
        primary_gr = "Not Applicable"

    return {
        "Primary_HL": primary_hl,
        "Support_HL": support_hl,
        "Lifestyle_HL": lifestyle_hl,
        "Primary_DF": primary_df,
        "Support_DF": support_df,
        "Lifestyle_DF": lifestyle_df,
        "Primary_GR": primary_gr
    }

# 4. Inject small noise to support/lifestyle products
def inject_noise(product_name, column_name):
    noise_options = {
        "Support_HL": [
            "Red Onion Oil", "Fenugreek Scalp Serum", "Hairfall Repair Combo"
        ],
        "Lifestyle_HL": [
            "Stress Yoga", "Scalp Steam", "Silk Pillowcase"
        ],
        "Support_DF": [
            "Neem Oil", "Anti-Dandruff Hair Mask", "Charcoal Shampoo"
        ],
        "Lifestyle_DF": [
            "Oiling Schedule", "Mild Conditioner", "Herbal Rinse"
        ]
    }
    return random.choice(noise_options[column_name])

# 5. Generate full dataset
n_rows = 28000
data = []
for _ in tqdm(range(n_rows)):
    user = generate_user_response()
    products = map_products(user)

    # Introduce noise randomly in support/lifestyle (5% chance)
    for key in products:
        if "Support" in key or "Lifestyle" in key:
            if products[key] != "Not Applicable" and random.random() < 0.05:
                products[key] = inject_noise(products[key], key)

    user.update(products)
    data.append(user)

# 6. Save to CSV
df = pd.DataFrame(data)
df.to_csv("responses_with_targets_noisy.csv", index=False)
print("✅ File saved as: responses_with_targets_noisy.csv")
```

```python
import pandas as pd
import random
from tqdm import tqdm
import numpy as np

# 1. Define feature space
features = {
    "AgeGroup": ["18–29", "30–44", "45+"],
    "HairLoss": ["Yes", "No"],
    "Dandruff": ["Yes", "No"],
    "HairGreying": ["Yes", "No"],
    "HL_Duration": ["<3 months", "3–6 months", ">6 months"],
    "HL_Shedding": ["Rarely", "Occasionally", "Frequently"],
    "DF_Type": ["Powdery", "Flaky"],
    "DF_ItchyScalp": ["Rarely", "Frequently"],
    "GR_Duration": ["<3 months", "<1 year", "<1.5 years"],
    "GR_TreatmentType": ["Coloring/Keratin", "None"],
    "RegularUsage": ["Yes", "No"],
    "WashFrequency": ["Rarely", "2–3 times/week", "Frequently"],
    "TakesSupplements": ["Yes", "No"],
    "StressLevel": ["Rarely", "Occasionally", "Frequently"],
    "FamilyHistory": ["Yes", "No"]
}

# 2. Support/Lifestyle noise options
noise_options = {
    "Support_HL": ["Red Onion Oil", "Fenugreek Scalp Serum", "Hairfall Repair Combo"],
    "Lifestyle_HL": ["Stress Yoga", "Scalp Steam", "Silk Pillowcase"],
    "Support_DF": ["Neem Oil", "Anti-Dandruff Hair Mask", "Charcoal Shampoo"],
    "Lifestyle_DF": ["Oiling Schedule", "Mild Conditioner", "Herbal Rinse"],
    "Primary_HL": ["Amla Hair Tonic", "Biotin Spray"],
    "Primary_DF": ["Apple Cider Rinse", "Coal Tar Shampoo"],
    "Primary_GR": ["Henna Oil", "Curry Leaf Capsules"]
}

# 3. Sample generation
def generate_user_response():
    return {k: random.choice(v) for k, v in features.items()}

# 4. Product mapping with slight randomness
def map_products(user):
    # Hair Loss
    if user["HairLoss"] == "Yes":
        duration = user["HL_Duration"]
        if duration == "<3 months":
            primary_hl = random.choice(["Redensyl + Biotin Combo Serum", "Redensyl Only"])
            support_hl = "Shampoo + Conditioner for Hairfall"
        elif duration == "3–6 months":
            primary_hl = "Redensyl + Biotin + Mild Minoxidil (Pilgrim, Be Bodywise)"
            support_hl = "DHT Blocker Shampoo + Conditioner"
        else:
            primary_hl = "Minoxidil 5% Foam + Redensyl + Microneedling"
            support_hl = "DHT Blocker + Supplements"
        lifestyle_hl = "Scalp Massage, Diet Improvement"
    else:
        primary_hl = support_hl = lifestyle_hl = "Not Applicable"

    # Dandruff
    if user["Dandruff"] == "Yes":
        if user["DF_Type"] == "Powdery" and user["DF_ItchyScalp"] == "Frequently":
            primary_df = "Ketoconazole + Salicylic Acid Shampoo"
            support_df = "Tea Tree Scalp Serum"
        else:
            primary_df = "ZPTO + pH Balancing Anti-Dandruff Shampoo"
            support_df = "Aloe Vera + Niacinamide Serum"
        lifestyle_df = "Avoid Hot Water, Gentle Shampoo"
    else:
        primary_df = support_df = lifestyle_df = "Not Applicable"

    # Greying
    if user["HairGreying"] == "Yes":
        if user["GR_TreatmentType"] == "Coloring/Keratin":
            primary_gr = "Hair Strengthening Oil + Amla Tablets"
        elif user["GR_Duration"] == "<3 months":
            primary_gr = "Vitamin B12 + Biotin Gummies"
        elif user["GR_Duration"] == "<1 year":
            primary_gr = "Melanin Boost Hair Oil"
        else:
            primary_gr = "Anti-Grey Hair Serum + Vitamin Supplements"
    else:
        primary_gr = "Not Applicable"

    return {
        "Primary_HL": primary_hl,
        "Support_HL": support_hl,
        "Lifestyle_HL": lifestyle_hl,
        "Primary_DF": primary_df,
        "Support_DF": support_df,
        "Lifestyle_DF": lifestyle_df,
        "Primary_GR": primary_gr
    }

# 5. Inject random noise
def add_random_noise(products, noise_chance=0.10):
    for col in products:
        if products[col] != "Not Applicable" and random.random() < noise_chance:
            if col in noise_options:
                products[col] = random.choice(noise_options[col])
    return products

# 6. Add mislabels to simulate human error (~2%)
def introduce_mislabel(products):
    if random.random() < 0.02:
        target_cols = [k for k in products if products[k] != "Not Applicable"]
        if target_cols:
            col = random.choice(target_cols)
            if col in noise_options:
                products[col] = random.choice(noise_options[col])
    return products

# 7. Dataset creation
n_rows = 28000
data = []

for _ in tqdm(range(n_rows)):
    user = generate_user_response()

    # Flip feature to simulate real-world data inconsistency (~2%)
    if random.random() < 0.02:
        flip_feature = random.choice(list(features.keys()))
        user[flip_feature] = random.choice(features[flip_feature])

    products = map_products(user)
    products = add_random_noise(products)
    products = introduce_mislabel(products)

    user.update(products)
    data.append(user)

# 8. Save
df = pd.DataFrame(data)
df.to_csv("responses_with_targets_noisy.csv", index=False)
print("✅ File saved as: responses_with_targets_noisy.csv")
```
